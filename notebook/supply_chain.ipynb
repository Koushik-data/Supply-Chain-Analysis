```python
null_report = df.isnull().sum()
print(null_report[null_report > 0]) 
```

```python
df = df[[
    'Order Id',
    'order date (DateOrders)',
    'shipping date (DateOrders)',
    'Sales',
    'Order Item Quantity',
    'Order Profit Per Order',
    'Category Name',
    'Product Name',
    'Days for shipping (real)',
    'Days for shipment (scheduled)',
    'Delivery Status',
    'Late_delivery_risk'
]]
```

```python
df.columns = [
    'order_id','order_date','shipping_date','sales','quantity',
    'profit','category','product','actual_shipping_days',
    'scheduled_shipping_days','delivery_status','late_risk'
]
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df['shipping_date'] = pd.to_datetime(df['shipping_date'])
```

```python
df.isnull().sum()
df = df.dropna()
```

```python
df = df.drop_duplicates()
```

```python
df.head()
```

```python
df['shipping_delay'] = df['actual_shipping_days'] - df['scheduled_shipping_days']
df['month'] = df['order_date'].dt.to_period('M')
```

```python
df['month'] = df['month'].astype(str)
```

```python
df.to_csv('clean_supply_chain.csv',index=False)
```
